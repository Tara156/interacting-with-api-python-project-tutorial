```python
for track in results['tracks']:
    print("Track:", track['name'])  # nombre de la canción
    print("Álbum:", track['album']['name'])  # nombre del álbum
    print("Artista principal:", track['artists'][0]['name'])  # nombre primer artista
    print("Link de Spotify:", track['external_urls']['spotify'])  # link canción
    print("-" * 30)
```

```python
# Solo imprimir el nombre de cada canción
for track in results['tracks']:
    print(track['name'])
```

```python
for track in results['tracks']:
    nombre = track['name']
    duracion_ms = track['duration_ms']  # duración en milisegundos
    duracion_min = duracion_ms / (1000 * 60)  # conversión directa a minutos
    popularidad = track['popularity']

    print(f"Canción: {nombre}")
    print(f"Duración: {duracion_min} minutos")  # sin formato de decimales
    print(f"Popularidad: {popularidad}/100")
    print("-" * 30)
```

```python
# Crear listas vacías
nombres = []
duraciones = []
popularidades = []

# Rellenar las listas
for track in results['tracks']:
    nombres.append(track['name'])
    duraciones.append(track['duration_ms'] / (1000 * 60))  # en minutos
    popularidades.append(track['popularity'])
```
